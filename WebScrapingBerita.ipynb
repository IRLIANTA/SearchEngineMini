````python
from wordcloud import WordCloud
````

````python
# Gabungkan semua teks menjadi satu dan ubah menjadi huruf kecil
all_text = ' '.join(df['Text'].tolist()).lower()
````

````python
# Hapus tanda baca dan pisahkan kata
words = [word.strip(string.punctuation) for word in all_text.split()]
````

````python
# Download the stopwords dataset if not already present
try:
    stopwords.words('indonesian')
except LookupError:
    print("Downloading NLTK stopwords...")
    nltk.download('stopwords')
    print("Download complete.")
````

````python
# Hapus stopwords dan kata kosong
stop_words = set(stopwords.words('indonesian'))
filtered_words = [w for w in words if w not in stop_words and w != '']
````

````python
# Ambil 10 kata paling sering muncul, lalu ubah ke huruf besar
common_words = Counter(filtered_words).most_common(10)
top_words_dict = {word.upper(): count for word, count in common_words}
````

````python
# Buat Word Cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='plasma'
).generate_from_frequencies(top_words_dict)
````

````python
# Tampilkan Word Cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud: 10 Kata Paling Sering Muncul (Huruf Besar)", fontsize=16)
plt.show()
````

````python
# Cetak hasil
print("\n10 Kata paling sering muncul (tanpa stopwords dan dalam huruf besar):")
for word, count in top_words_dict.items():
    print(f"{word}: {count} kali")
````
